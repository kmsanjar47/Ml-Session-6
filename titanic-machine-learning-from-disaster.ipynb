```python
X_train = pd.read_csv("/kaggle/input/titanic/train.csv")
y_test = pd.read_csv("/kaggle/input/titanic/test.csv")
```

```python
X_train.head(5)
```

```python
#Dropping Fare,Ticket,Name from Train
X_train.drop(["PassengerId","Name","Ticket","Fare"],axis = 1, inplace = True)
y_test_result = y_test["PassengerId"]
y_test.drop(["PassengerId","Name","Ticket","Fare"],axis = 1, inplace = True)
```

```python
X_test = X_train["Survived"]
X_train.drop("Survived",axis = 1, inplace = True)
```

```python
X_test
```

```python
#Handling null values
X_train.isna().sum()
```

```python
y_test.isna().sum()
```

```python
X_train["Age"].fillna(value = X_train["Age"].mean(),inplace = True)
y_test["Age"].fillna(value = y_test["Age"].mean(),inplace = True)
```

```python
X_train["Cabin"].fillna(value = X_train["Cabin"].mode().iloc[0],inplace = True)
X_train["Embarked"].fillna(value = X_train["Embarked"].mode().iloc[0],inplace = True)
y_test["Cabin"].fillna(value = y_test["Cabin"].mode().iloc[0],inplace = True)
y_test["Embarked"].fillna(value = y_test["Embarked"].mode().iloc[0],inplace = True)
```

```python
X_train.isna().sum()
```

```python
y_test.isna().sum()
```

```python
#Encoding Sex, Cabin, Embarked
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
X_train["Sex"] = le.fit_transform(X_train[["Sex"]])
X_train["Cabin"] = le.fit_transform(X_train[["Cabin"]])
X_train["Embarked"] = le.fit_transform(X_train[["Embarked"]])
```

```python
y_test["Sex"] = le.fit_transform(y_test[["Sex"]])
y_test["Cabin"] = le.fit_transform(y_test[["Cabin"]])
y_test["Embarked"] = le.fit_transform(y_test[["Embarked"]])
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train,X_test)
```

```python
y_predicted = lr.predict(y_test)
```
